```python
import numpy as np
import pandas as pd
import nibabel as nib
import matplotlib.pyplot as plt

from IPython.display import clear_output
```

```python
SUBJECT = "sub-01"
SCENARIO = "perceptionTraining01"
DATA_PATH = f"data/ds001246/derivatives/preproc-spm/output/{SUBJECT}/ses-{SCENARIO}/func"
# DATA_PATH = f"data/ds001246/{SUBJECT}/ses-{SCENARIO}/func"
TARGETS_PATH = f"data/ds001246/{SUBJECT}/ses-{SCENARIO}/func"
```

```python
TASK = "task-perception_run-01"

nii_data = nib.load(f"{DATA_PATH}/{SUBJECT}_ses-{SCENARIO}_{TASK}_bold_preproc.nii.gz").get_fdata()
targets = pd.read_csv(f"{TARGETS_PATH}/{SUBJECT}_ses-{SCENARIO}_{TASK}_events.tsv", sep="\t")
```

```python
nii_data.shape
```

```python
# visualize data
for t_point in range(0, nii_data.shape[3]):
    clear_output(wait=True)

    plt.figure(figsize=(8, 8))

    for slice_num in range(0, nii_data.shape[2]):
        plt.subplot(7, 8, slice_num + 1)
        plt.imshow(nii_data[:, :, slice_num, t_point], cmap="seismic")
        plt.axis("off")

    plt.suptitle(f"t = {t_point}")
    plt.tight_layout()
    plt.show()
```
