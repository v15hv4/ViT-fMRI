```python
import seaborn as sns
import matplotlib.pyplot as plt

from data import GODData
```

```python
sns.set_style("whitegrid")
```

```python
rois = ["V1d", "V1v", "V2d", "V2v", "V3d", "V3v", "hV4", "LOC", "FFA", "PPA", "HVC"]
train_dataset = GODData(subject="01", rois=rois)
```

```python
for f, l in train_dataset:
    plt.figure(figsize=(10, 10))
    for slidx in range(f.shape[1]):
        plt.subplot(8, 8, slidx + 1)
        plt.imshow(f[1, slidx, :, :], cmap="seismic")
        plt.axis("off")
    plt.tight_layout(pad=0.5)
    plt.show()
    break
```
